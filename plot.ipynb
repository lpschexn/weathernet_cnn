```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Precision and Recall values for DROR, DSOR, and WeatherNet CNN
name = ['DROR', 'WeatherNet CNN', 'DSOR']
precision = [71.51, 96.35, 65.07]
recall = [91.89, 78.64, 95.60]

# Bar plot configuration
bar_width = 0.25
xlim_spacing = 0.9

fig, ax = plt.subplots()

br1 = np.arange(len(precision))
br2 = [x + bar_width + bar_width/2 for x in br1]
br3 = [x + bar_width + bar_width/2 for x in br2]

br1 = br1 + bar_width/2
plt.bar(br1, precision, width=bar_width, label='Precision')
plt.bar(br2, recall, width=bar_width, label='Recall')

plt.xticks([r + bar_width for r in range(len(precision))], name)

for bar in ax.containers:
    ax.bar_label(bar)



plt.legend(loc='lower right')
plt.title('Precision and Recall of DSOR, WeatherNet CNN, and DSOR')
plt.ylabel('Percentage [%]')
plt.ylim((40, 100))
plt.show()
```
